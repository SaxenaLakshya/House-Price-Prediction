## Scikit-learn Design

Primarily, three types of objects
1. Estimators - It estimates some parameters based on a dataset. Eg. imputer
It has a fit method and transform method.
Fit Method - Fits the dataset and calculates internal parameters

2. Transformers - transform method taked input and returns output based on the learnings from fit().
It also has convenience function called fit_transform() which fits and then transforms.

3. Predictors - LinearRegression model is an example of predictor. fit() and predict()
are two common functions. It also gives score() function which will evaluate the predictions

## Feature Scaling

Primarily, two types of feature scaling methods:
1. Min-max scaling (Normalization)
(value - min) / (max - min)
Sklearn provides a class called MinMaxScaler for this

2. Standardization
(value - mean) / std
Sklearn provides and class called Standard Scaler for this

## Creating a Pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
my_pipeline = Pipeline([
    ('imputer',SimpleImputer(strategy="median")),
    # ...... add as many you want in your Pipeline
    ('std_scaler',StandardScaler()),
])
```

```python
housing_num_tr = my_pipeline.fit_transform(housing)
housing_num_tr.shape
```

## Selecting a Desired Model House Price Predition

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
# model = LinearRegression()
# model = DecisionTreeRegressor()
model = RandomForestRegressor()
model.fit(housing_num_tr, housing_labels)
```

```python
some_data = housing.iloc[0:5]
some_labels = housing_labels.iloc[:5]
prepared_data = my_pipeline.transform(some_data)
```

```python
model.predict(prepared_data)
```

```python
list(some_labels)
```

## Evaluating the Model

```python
from sklearn.metrics import mean_squared_error
import numpy as np
housing_predictions = model.predict(housing_num_tr)
mse = mean_squared_error(housing_labels, housing_predictions)
rmse = np.sqrt(mse)
```

```python
rmse
```

## Using better Evaluation Technique - Cross Validation

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model, housing_num_tr, housing_labels, scoring="neg_mean_squared_error", cv=10)
rmse_score = np.sqrt(-scores)
```

```python
rmse_score
```

```python
def print_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard Deviation:",scores.std())
```

```python
print_scores(rmse_score)
```

## Testing the Model on Test Data

```python
X_test = strat_test_set.drop("MEDV", axis=1)
Y_test = strat_test_set["MEDV"].copy()
X_test_prepared = my_pipeline.transform(X_test)
final_predictions = model.predict(X_test_prepared)
final_mse = mean_squared_error(Y_test, final_predictions)
final_rmse = np.sqrt(final_mse)
```
